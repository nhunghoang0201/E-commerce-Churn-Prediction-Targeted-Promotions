## **2.2. Preprocessing**

### **2.2.1.Change datatype**

```python
churn_predict["CustomerID"]=churn_predict["CustomerID"].astype("string")
churn_predict["Tenure"] = churn_predict["Tenure"].round().astype("Int64")
churn_predict["HourSpendOnApp"]=churn_predict["HourSpendOnApp"].astype("Int64")
churn_predict["CouponUsed"]=churn_predict["CouponUsed"].astype("Int64")
churn_predict["OrderCount"]=churn_predict["OrderCount"].astype("Int64")
```

### **2.2.2. Null Value**

**Process Tenure null value**

```python
Null_Tenure=churn_predict[churn_predict["Tenure"].isnull()]
print(Null_Tenure["HourSpendOnApp"].unique())
print(Null_Tenure["OrderCount"].unique())
```

When Tenure is null, HourSpendOnApp & OrderCount are more than 0 and are not null, which means that Customer has logged in and placed one or more orders => Tenure is null due to error in data entry

=>  Missing Tenure should imputed using the median tenure grouped by churn status

```python
churn_predict["Tenure"] = churn_predict.groupby("Churn")["Tenure"].transform(
    lambda x: x.fillna(x.median()))
```

**Process HourSpendOnApp null value**

```python
Null_HourSpendOnApp=churn_predict[churn_predict["HourSpendOnApp"].isnull()]
print(Null_HourSpendOnApp["OrderCount"].unique())
```

When HourSpendOnApp is null, OrderCount are more than 0 and are not null, which means that Customer placed one or more orders => HourSpendOnApp is null due to error in data entry

=>  Missing HourSpendOnApp should imputed using the median tenure grouped by PreferredLoginDevice

```python
churn_predict["HourSpendOnApp"] = churn_predict.groupby("PreferredLoginDevice")["HourSpendOnApp"].transform(
    lambda x: x.fillna(x.median()))
```

**Process WarehouseToHome null values**

HourSpendOnApp is null due to error in data entry

=> HourSpendOnApp should imputed using the median tenure grouped by CityTier

```python
churn_predict["WarehouseToHome"]=churn_predict.groupby("CityTier")["WarehouseToHome"].transform(lambda x: x.fillna(x.median()))
```

**Process OrderAmountHikeFromlastYear null values**

```python
Null_OrderAmountHikeFromlastYear=churn_predict[churn_predict["OrderAmountHikeFromlastYear"].isnull()]
print(Null_OrderAmountHikeFromlastYear["Tenure"].unique())
print(Null_OrderAmountHikeFromlastYear["DaySinceLastOrd"].unique())
```

```python
churn_predict["OrderAmountHikeFromlastYear"] = churn_predict.groupby("OrderCount")["OrderAmountHikeFromlastYear"].transform(
    lambda x: x.fillna(x.median()))
```

**Process CouponUsed**

```python
churn_predict["CouponUsed"] = churn_predict.groupby("OrderCount")["CouponUsed"].transform(
    lambda x: x.fillna(x.median()))
```

**Process OrderCount**

```python
churn_predict["OrderCount"] = churn_predict.groupby("Tenure")["OrderCount"].transform(
    lambda x: x.fillna(x.median()))
```

**Process DaySinceLastOrder**

```python
churn_predict["DaySinceLastOrder"] = churn_predict.groupby("Tenure")["DaySinceLastOrder"].transform(
    lambda x: x.fillna(x.median()))
```

# **3. Feature Engineering**

## **3.1. Categorical Encoding**

```python
churn_predict_cat = churn_predict.select_dtypes(include='object').columns
encoded = pd.get_dummies(churn_predict[churn_predict_cat], drop_first=True)
encoded = encoded.astype(int)
churn_predict_enc = pd.concat([churn_predict.drop(columns=churn_predict_cat), encoded], axis=1)
churn_predict_enc.head(5)
```

## **3.2. Dimension Reduction**

```python
pca = PCA(n_components=3)
pca.fit(churn_predict_enc)
PCA_ds = pd.DataFrame(pca.transform(churn_predict_enc), columns=(["col1","col2", "col3"]))
PCA_ds
```

```python
pca.explained_variance_ratio_
```

# **4. Model Training**

## **4.1. Apply KMeans cluster**

### **4.1.1. Choosing K Means**

```python
ss = []
max_clusters = 10
for i in range(1, max_clusters+1):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(PCA_ds)
    # Inertia method returns WCSS for that model
    ss.append(kmeans.inertia_)

# Plot the Elbow method
plt.figure(figsize=(10,5))
plt.plot(range(1, max_clusters+1), ss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## **4.1.2. Apply K Means**

```python
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
predicted_labels = kmeans.fit_predict(PCA_ds)
PCA_ds['clusters']=predicted_labels
churn_predict['clusters']=predicted_labels
churn_predict
```

## **4.2. Evaluating Model**

### **4.2.1. Silhouette Score**

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(PCA_ds, predicted_labels)
print(sil_score)
```
